```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = "Best_State_Project_Files/Average_Salary.csv"
```

```python
file_df = pd.read_csv(file)
file_df.head()
```

```python
x_axis = np.arange(0, len(file_df))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)



#fig = plt.gcf()
#fig.set_size_inches(28.5, 20.5)
#plt.figure(figsize=(30,20)) 
#salary_plot = file_df.plot(kind="bar", title="Average Annual Salary", align="center")
figsize_x = 10
figsize_y = 3

figsize, ax1 = plt.subplots(figsize=(figsize_x, figsize_y))

salary_plot = ax1.bar(file_df["State Long"],
                     file_df["Annual Wages per Employee"],
                     align="center", alpha=0.5, color="r")
ax1.set_xlabel("State")
ax1.set_ylabel("Income")
ax1.set_title("Average Income per State")
plt.xlim(-1, 51)
plt.xticks(tick_locations, file_df["State Long"], rotation='vertical', fontsize=10)

plt.savefig("Best_State_Project_Files/Average_Salary.PNG")

#plt.tight_layout()
```
